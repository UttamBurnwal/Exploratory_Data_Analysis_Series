**Use Pandas to store given data in DataFrames and use Matplotlib or Plotly to Plot Bitcoin’s ‘Low‘ and its corresponding ‘Quantum’ for the first and last month of trade data available. Quantum = Difference in low for two consecutive days.**


```python
bitcoin.head()
```

```python
bitcoin.tail()
```

**Date column holds both date and time so I am creating two seperate columns for date and time**

```python
def seperate_date_time(dataframe, column):
  dataframe["date"] = pd.to_datetime(dataframe[f'{column}']).dt.date
  dataframe["time"] = pd.to_datetime(dataframe[f'{column}']).dt.time
  dataframe.drop([f'{column}'], axis = 1, inplace = True)
```

```python
seperate_date_time(bitcoin, "Date")
```

## First Month

The very month in the data was of April however the data for that month was incomplete and included values of only last two days. So I am taking May as the first month.

```python
bitcoin_first_month = bitcoin[bitcoin['date'].map(lambda x: (x.year == 2013) & (x.month == 5))]
bitcoin_first_month.head(1)
```

```python
# Function for the difference in low for two consecutive days

def diff_low(day):
  return day.iloc[-1] - day.iloc[0]
```

```python
bitcoin_first_month['quantum'] = bitcoin_first_month['Low'].rolling(2).apply(diff_low)
```

```python
if not os.path.exists("Task1"):
    os.mkdir("Task1")
```

```python
plt.figure(figsize = (10,8))
plt.plot(bitcoin_first_month['date'],bitcoin_first_month['quantum'], 'go-', label='Quantum')
plt.plot(bitcoin_first_month['date'],bitcoin_first_month['Low'], 'rs-', label='Low')
plt.xticks(rotation = 'vertical')
plt.legend(loc='best')
plt.title('Distribution of Low and Quantum Values for the First month')
plt.xlabel('Date')
plt.ylabel('Value')
for x1, y1 in zip(bitcoin_first_month['date'],bitcoin_first_month['Low']):
  plt.annotate(y1, (x1,y1), rotation = 90, size = 10)
for x2, y2 in zip(bitcoin_first_month['date'],bitcoin_first_month['quantum']):
  plt.annotate(y2, (x2,y2), rotation = 90, size = 10)
plt.savefig("Task1/Bitcoin_First_Month.pdf")
plt.show()
```

## Last Month

The last month in the data was of July 2021 however the data for that month was incomplete and included values of only last six days. So I am taking June 2021 as the last month.

```python
bitcoin_last_month = bitcoin[bitcoin['date'].map(lambda x: (x.year == 2021) & (x.month == 6))]
```

```python
bitcoin_last_month['quantum'] = bitcoin_last_month['Low'].rolling(2).apply(diff_low)
```

```python
plt.figure(figsize = (10,8))
plt.plot(bitcoin_last_month['date'],bitcoin_last_month['quantum'], 'go-', label='Quantum')
plt.plot(bitcoin_last_month['date'],bitcoin_last_month['Low'], 'rs-', label='Low')
plt.xticks(rotation = 'vertical')
plt.legend(loc='best')
plt.title('Distribution of Low and Quantum Values for the Last month')
plt.xlabel('Date')
plt.ylabel('Value')
for x1, y1 in zip(bitcoin_last_month['date'],bitcoin_last_month['Low']):
  plt.annotate(y1, (x1,y1), rotation = 90, size = 10)
for x2, y2 in zip(bitcoin_last_month['date'],bitcoin_last_month['quantum']):
  plt.annotate(y2, (x2,y2), rotation = 90, size = 10)
plt.savefig("Task1/Bitcoin_Last_Month.pdf")
plt.show()
```
